```python
import matplotlib.pyplot as plt
his_plt = train_data[['state', 'cfips']].drop_duplicates(subset=['cfips']).groupby(['state']).size().sort_values()

fig, ax = plt.subplots(figsize=(10,14))
bars = ax.barh(his_plt.index, his_plt.values)
ax.bar_label(bars)
ax.set_title('Number of counties per state')
plt.show()
```
